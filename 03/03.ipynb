# 03 데이터 분석에 필요한 연장 챙기기  

## 03-1 변수  

### 변수(variable)  
- 변수: 다양한 값을 지닌 하나의 속성  
- 데이터 분석의 대상  
- 데이터 분석은 변수 간에 어떤 관계가 있는지 파악하는 작업  

### 상수(constant)  
- 상수: 하나의 값으로만 되어 있는 속성  
- 분석 대상이 될 수 없음  

## 03-2 함수  
- 데이터 분석은 함수를 이용해서 변수를 조작하는 일  
- 함수에 값을 넣으면 특정한 기능을 수행해 처음과 다른 값으로 만듦  

## 03-3 패키지  
### 패키지(package)  
- 패키지에는 다양한 함수가 들어 있음  

```python
import seaborn as sns
```

```python
# load_dataset() : Get access to an example dataset
df = sns.load_dataset('titanic')
df
```

```python
# countplot() : Show the counts of observations in each categorical bin using bars
sns.countplot(data=df, x='class')
```

```python
# hue : 변수의 항목별로 막대의 색을 다르게 표현
sns.countplot(data=df, x='class', hue='alive')
```

```python
# Plot horizontally to make more space for category labels
sns.countplot(data=df, y='class', hue='alive')
```
