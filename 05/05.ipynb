## 05-2 변수명 바꾸기
- pandas.DataFrame.rename() : Alter axes labels

```python
df_temp = pd.DataFrame({'var1' : [1, 2, 3], 'var2' : [4, 5, 6]})
df_temp
```

```python
# pandas.DataFrame.copy(deep=True) : Make a copy of this object’s indices and data
# When deep=True (default), a new object will be created with a copy of the calling object’s data and indices. 
# Modifications to the data or indices of the copy will not be reflected in the original object (see notes below).
# When deep=False, a new object will be created without copying the calling object’s data or index (only references to the data and index are copied). 
# Any changes to the data of the original will be reflected in the shallow copy (and vice versa).

df_new = df_temp.copy()
df_new
```

```python
# pandas.DataFrame.rename()
df_new = df_new.rename(columns = {'var2' : 'v2'})    # Alter var2 to v2
df_new
```

## 05-3 파생변수 만들기

```python
df = pd.DataFrame({'var1' : [4, 3, 8], 'var2' : [2, 6, 1]})
df
```

```python
df['var_sum'] = df['var1'] + df['var2']
df
```

```python
df['var_mean'] = df['var_sum'] / 2
df
```

### 조건문을 활용해 파생변수 만들기
mpg 통합 연비 변수(total)이 기준을 충족해 고연비 합격 판정을 받은 자동차가 몊 대나 되는지 알아보기.

1. total = (cty + hwy) / 2

```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

2. 기준값 정하기  
total의 연비와 중앙값이 약 20  
total이 20~22.5 사이인 자동차 모델이 가장 많고, 대부분 25 이하  
따라서 기준값을 20으로 설정  

```python
mpg['total'].describe()
```

```python
# pandas.DataFrame.hist() : Make a histogram of the DataFrame’s columns
mpg['total'].plot.hist()
```

3. 합격 판정 변수(test) 만들기

```python
# numpy.where(condition, [x, y, ]/) : Return elements chosen from x or y depending on condition
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')    # If mpg['total'] >= 20, then mpg['test'] = 'pass', else mpg['test'] = 'fail'
mpg.head()
```

4. 빈도표로 합격 판정 자동차 수 알아보기

```python
# pandas.DataFrame.value_counts() : Return a Series containing counts of unique rows in the DataFrame
mpg['test'].value_counts()
```

5. 막대그래프로 빈도 표현하기

```python
test_cnt = mpg['test'].value_counts()    # 연비 합격 빈도표를 변수에 할당

# pandas.DataFrame.plot.bar() : Vertical bar plot
test_cnt.plot.bar()
```

```python
test_cnt.plot.bar(rot=0)    # 축 이름 수평으로 만들기
```

### 중첩조건문  
연비 등급을 total 값이 30 이상이면 A, 20 이상 30 미만이면 B, 20 미만이면 C로 분류하여 연비 등급 변수(grade) 만들기

```python
mpg['grade'] = np.where(mpg['total'] >= 30, 'A', 
                        np.where(mpg['total'] >= 20, 'B', 'C'))
mpg.head()
```

```python
grade_cnt = mpg['grade'].value_counts()
grade_cnt
```

```python
grade_cnt.plot.bar(rot=0)
```

```python
grade_cnt = grade_cnt.sort_index()    # Sort index in alphabetical order
grade_cnt
```

```python
grade_cnt.plot.bar(rot=0)
```

### 목록에 해당하는 행으로 변수 만들기
mpg의 size에 categroy가 compact, subcompact, 2seater면 small을, 아니면 large를 부여

```python
mpg['size'] = np.where((mpg['category'] == 'compact') 
                       | (mpg['category'] == 'subcompact') 
                       | (mpg['category'] == '2seater'), 'small', 'large')
# mpg['size'] = np.where(mpg['category'].isin(['compact', 'subcompact', '2seater']), 'small', 'large')
size_cnt = mpg['size'].value_counts()
size_cnt
```
