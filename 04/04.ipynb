# 04 데이터 프레임의 세계로!  

## 04-1 데이터 프레임  

### 데이터 프레임(data frame)  
- 데이터를 다룰 때 가장 많이 사용하는 데이터 형태  
- 행과 열로 구성된 사각형 모양의 표 꼴  

- 열은 속성(column, variable)  
- 행은 정보(row, case)  

- 데이터가 크다 = 행이 많다 or 열이 많다  
- 열이(변수가) 많아지면, 적용할 분석 기술이 달라지므로 데이터의 다양성을 의미하는 열이 많은 것이 더 중요

## 04-2 데이터 프레임 만들기
- pandas.DataFrame() : Two-dimensional, size-mutable, potentially heterogeneous tabular data

```python
import pandas as pd

# pandas.DataFrame()
df = pd.DataFrame({'name'   : ['Anthony', 'Benedict', 'Collin', 'Daphne'],
                  'english' : [90, 80, 60, 70],
                  'math'    : [50, 60, 100, 20]})
df
```

- 특정 변수의 값 활용

```python
# 값 추출
df['english']    # df의 english에 해당하는 값만 추출
```

```python
# 합
sum(df['english'])
```

### 혼자서 해보기

```python
# Q1
df = pd.DataFrame({'제품' : ['사과', '딸기', '수박'],
                   '가격' : [1800, 1500, 3000],
                   '판매량' : [24, 38, 13]})
df
```

```python
# Q2
print(f"과일 가격 평균 : {sum(df['가격']) / 3}")
print(f"판매량 평균 : {sum(df['판매량']) / 3}")
```

## 04-3 외부 데이터 이용

### Excel File
- read_excel() : Read an Excel file into a pandas DataFrame

```python
# read_excel()
df_exam = pd.read_excel('../data/excel_exam.xlsx')
df_exam
```

```python
# Compute average using len()
# len(dataframe) returns the number of rows of dataframe

sum(df_exam['english']) / len(df_exam)    # 영어 점수 평균
```

- read_excel()은 엑셀 파일의 첫 행을 변수명으로 인식함
- 만약 첫 행이 변수명이 아니라면 header=None parameter를 사용하여 오류 방지

```python
# pandas.DataFrame.read_excel(file, header=None)
df_exam_novar = pd.read_excel('../data/excel_exam_novar.xlsx', header=None)
df_exam_novar
```

- 엑셀 파일에 시트가 여러 개 있다면 sheet_name parameter 설정

```python
# Read data of Sheet2
df_exam = pd.read_excel('../data/excel_exam.xlsx', sheet_name='Sheet2')

# Read data of third sheet
df_exam = pd.read_excel('../data/excel_exam.xlsx', sheet_name=2)
```

### CSV File  
- Comma-Separated Values  
- read_csv() : Read a CSV file into DataFrame

```python
# read_csv()
df_exam = pd.read_csv('../data/exam.csv')
df_exam
```

### Write CSV File
- pandas.DataFrame.to_csv() : Write object to a CSV file

```python
# pandas.DataFrame()
df_midterm = pd.DataFrame({'english' : [90, 80, 60, 70],
                           'math'    : [50, 60, 100, 20],
                           'nclass'  : [1, 1, 2, 2]})
df_midterm
```

```python
# pandas.DataFrame.to_csv() 
df_midterm.to_csv("midterm.csv")
```
